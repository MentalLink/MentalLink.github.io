## - Audio to Text
Used a whisper model that utilses Dynamic Time Warping (DTW) applied to cross-attention weights to provide probabilites of timestamps for EACH word, this solves the issue of alignment.

We can then focus on chunking. Since we were told its a semantic chunking task, I avoided using LLMs to prompt the text data in and generate meaningful short clips from the text. Instead, I used purely semantic similarities to capture the chunks.

Used Cohere's free embeddings for this task.

Once the whisper (large) provided the text and its timestamps, the task became identifying which of the segments to focus on. So, I used the regular tokenizing that whisper uses and outputs as my base sentences as on average they were 4-5 seconds long. this meant i could use 3 of these to form a chunk which is desirable for me. I then grouped all the small parts into chunks of three and created embeddings for them. I then proceeded to take cosine similarities of each of these chunks of three with the previous, proceeding and whole text embeddings. This gave me a measure of a) how relevant a chunk is to its neighbours and b) how relevant it is to the whole text. Filtering from this, I was able to generate areas of high meaning and distinctness, which I used for my measure of success. I will attach a result I had gotten below this cell for reference, but the code should be able to run. In ideality, after those chunks were filtered by 95% percentile, we could group them >15s and creat meaningful audio and text pairs that dont lose meaning, at all and stand independtly.

I then proceeded to make the grad.io interface which can input any youtube link and output the pairs. However, it takes 6/7 minutes for a whisper large model to process the youtube video, so that is a weakness with productinising on colab, but given the right backend, this is an achievable task.

We need to find some way to see purely based on meaning, what distinct meaningful segments are.
````python
def combine_sentences(sentences, buffer_size=1):
    # Go through each sentence dict
    for i in range(len(sentences)):

        # Create a string that will hold the sentences which are joined
        combined_sentence = ''

        # Add sentences before the current one, based on the buffer size.
        for j in range(i - buffer_size, i):
            # Check if the index j is not negative (to avoid index out of range like on the first one)
            if j >= 0:
                # Add the sentence at index j to the combined_sentence string
                combined_sentence += sentences[j]['text'] + ' '

        # Add the current sentence
        combined_sentence += sentences[i]['text']

        # Add sentences after the current one, based on the buffer size
        for j in range(i + 1, i + 1 + buffer_size):
            # Check if the index j is within the range of the sentences list
            if j < len(sentences):
                # Add the sentence at index j to the combined_sentence string
                combined_sentence += ' ' + sentences[j]['text']

        # Then add the whole thing to your dict
        # Store the combined sentence in the current sentence dict
        sentences[i]['combined_text'] = combined_sentence

    return sentences

l = combine_sentences(l, buffer_size=1)
````

````python
from sklearn.metrics.pairwise import cosine_similarity

def calculate_cosine_distances(sentences, fullembed):
    distances = []
    for i in range(len(sentences) - 1):
        embedding_current = sentences[i]['combined_text_embedding']
        #embedding_next = sentences[i + 1]['combined_text_embedding']
        embedding_next = [x + y for x, y in zip(sentences[i +1]['text_embedding'], fullembed[0])]
        #embedding_next = fullembed[0]
        # Calculate cosine similarity
        similarity = cosine_similarity([embedding_current], [embedding_next])[0][0]

        # Convert to cosine distance
        distance = 1 - similarity

        # Append cosine distance to the list
        distances.append(distance)

        # Store distance in the dictionary
        sentences[i]['distance_to_next'] = distance

    # Optionally handle the last sentence
    # sentences[-1]['distance_to_next'] = None  # or a default value

    return distances, sentences
````

````python
import numpy as np

plt.plot(distances);

y_upper_bound = .8
plt.ylim(0, y_upper_bound)
plt.xlim(0, len(distances))

# We need to get the distance threshold that we'll consider an outlier
# We'll use numpy .percentile() for this
breakpoint_percentile_threshold = 95
breakpoint_distance_threshold = np.percentile(distances, breakpoint_percentile_threshold) # If you want more chunks, lower the percentile cutoff
plt.axhline(y=breakpoint_distance_threshold, color='r', linestyle='-');

# Then we'll see how many distances are actually above this one
num_distances_above_theshold = len([x for x in distances if x > breakpoint_distance_threshold]) # The amount of distances above your threshold
plt.text(x=(len(distances)*.01), y=y_upper_bound/50, s=f"{num_distances_above_theshold + 1} Chunks");

# Then we'll get the index of the distances that are above the threshold. This will tell us where we should split our text
indices_above_thresh = [i for i, x in enumerate(distances) if x > breakpoint_distance_threshold] # The indices of those breakpoints on your list

# Start of the shading and text
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
for i, breakpoint_index in enumerate(indices_above_thresh):
    start_index = 0 if i == 0 else indices_above_thresh[i - 1]
    end_index = breakpoint_index if i < len(indices_above_thresh) - 1 else len(distances)

    plt.axvspan(start_index, end_index, facecolor=colors[i % len(colors)], alpha=0.25)
    plt.text(x=np.average([start_index, end_index]),
             y=breakpoint_distance_threshold + (y_upper_bound)/ 20,
             s=f"{i}", horizontalalignment='center',
             rotation='vertical')

# # Additional step to shade from the last breakpoint to the end of the dataset
if indices_above_thresh:
    last_breakpoint = indices_above_thresh[-1]
    if last_breakpoint < len(distances):
        plt.axvspan(last_breakpoint, len(distances), facecolor=colors[len(indices_above_thresh) % len(colors)], alpha=0.25)
        plt.text(x=np.average([last_breakpoint, len(distances)]),
                 y=breakpoint_distance_threshold + (y_upper_bound)/ 20,
                 s=f"Chunk #{i+1}",
                 rotation='vertical')

plt.title("PG Essay Chunks Based On Embedding Breakpoints")
plt.xlabel("Index of sentences in essay (Sentence Position)")
plt.ylabel("Cosine distance between sequential sentences")
plt.show()
````

## Some plots for chapter 1

````python
import os
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set the directory containing the audio files
audio_dir = '/content/new-test'

# Get a list of audio files


audiofiles = []
for f in os.listdir(audio_dir):
  if f.startswith("01") and f.endswith(".mp3"):
    audiofiles.append(f)


features = []

# Loop through the audio files
for audio_file in audiofiles[:20]:
    # Load the audio file
    audio_path = os.path.join(audio_dir, audio_file)
    y, sr = librosa.load(audio_path)

    # Extract audio features
    rms = np.mean(librosa.feature.rms(y=y))
    spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
    spectral_bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))
    rolloff = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr))
    zcr = np.mean(librosa.feature.zero_crossing_rate(y))
    mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr), axis=1)
    chroma = np.mean(librosa.feature.chroma_stft(y=y, sr=sr), axis=1)
    mel = np.mean(librosa.feature.melspectrogram(y=y, sr=sr))
    contrast = np.mean(librosa.feature.spectral_contrast(y=y, sr=sr))
    tonnetz = np.mean(librosa.feature.tonnetz(y=y, sr=sr))

    features.append([rms, spectral_centroid, spectral_bandwidth, rolloff, zcr] + list(mfcc) + list(chroma) + [mel, contrast, tonnetz])

# Convert the features list to a numpy array
features_array = np.array(features)

# Compute descriptive statistics
print("Descriptive Statistics:")
feature_names = ['RMS', 'Spectral Centroid', 'Spectral Bandwidth', 'Spectral Rolloff', 'Zero-Crossing Rate'] + [f'MFCC {i}' for i in range(20)] + [f'Chroma {i}' for i in range(12)] + ['Mel Spectrogram', 'Spectral Contrast', 'Tonnetz']
for i, name in enumerate(feature_names):
    print(f"{name}: Mean={np.mean(features_array[:, i])}, Std={np.std(features_array[:, i])}")

# Plot histograms
fig, axs = plt.subplots(4, 4, figsize=(16, 12))
axs = axs.flatten()
for i, ax in enumerate(axs):
    ax.hist(features_array[:, i], bins=20)
    ax.set_title(feature_names[i])
plt.tight_layout()
plt.show()

# Plot correlation matrix
corr_matrix = np.corrcoef(features_array.T)
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, xticklabels=feature_names, yticklabels=feature_names, annot=True, cmap="YlGnBu")
plt.title("Correlation Matrix")
plt.show()
````

````python
import re, string
transcript_dir = '/content/new-test'

# Get a list of audio files
# Initialize lists to store features
text_features = []

# Loop through the audio files and corresponding transcripts
for audio_file in audiofiles[:20]:
    # Load the transcript file
    transcript_path = os.path.join(transcript_dir, audio_file.replace('.mp3', '.txt'))
    with open(transcript_path, 'r') as f:
        text = f.read()

    # Extract text features
    num_chars = len(text)
    num_words = len(text.split())
    num_lines = len(text.split('\n'))
    num_unique_chars = len(set(text))
    avg_word_length = sum(len(word) for word in text.split()) / num_words if num_words > 0 else 0

    text_features.append([num_chars, num_words, num_lines, num_unique_chars, avg_word_length])

# Convert the text_features list to a numpy array
text_features_array = np.array(text_features)

# Compute descriptive statistics
print("Descriptive Statistics:")
feature_names = ['Number of Characters', 'Number of Words', 'Number of Lines', 'Number of Unique Characters', 'Average Word Length']
for i, name in enumerate(feature_names):
    print(f"{name}: Mean={np.mean(text_features_array[:, i])}, Std={np.std(text_features_array[:, i])}")

# Plot histograms
fig, axs = plt.subplots(1, 5, figsize=(20, 4))
for i, ax in enumerate(axs):
    ax.hist(text_features_array[:, i], bins=20)
    ax.set_title(feature_names[i])
plt.tight_layout()
plt.show()
````
